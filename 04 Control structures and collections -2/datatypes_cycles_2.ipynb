**Задание 2**. Выведите на экран все уникальные гео-ID из значений словаря ids. Т. е. список вида [213, 15, 54, 119, 98, 35]

```python
ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}
```

```python
idd = []

id = ids.values()
for el in id:
    for element in el: 
        if element not in idd:
            idd.append(element)
print(idd)
```

**Задание 3**. Дан список поисковых запросов. Получить распределение количества слов в них. 
Т. е. поисковых запросов из одного - слова 5%, из двух - 7%, из трех - 3% и т.д.

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
```

```python
len_all = len(queries)
new_queries = [new_queries.split(' ') for new_queries in queries]
count = [len(x) for x in new_queries]
# print(count)
count_set = set(count)
# print(count_set)

for len_ in count_set:
    print(f'Запрос с количеством слов {len_} встречается в {round(count.count(len_) / len_all * 100, 2)} % случаев')  
```

**Задание 4**. Дана статистика рекламных каналов по объемам продаж. Напишите скрипт, который возвращает название канала с максимальным объемом.
Т. е. в данном примере скрипт должен возвращать 'yandex'.

```python
stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98}
```

```python
stats_item = stats.items()
max_value = max(stats.values())
for stats_val in stats_item:
    if stats_val[1] == max_value:
        print(stats_val[0])
```
